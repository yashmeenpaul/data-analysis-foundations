```python
#When you have null values in the dataset, you can't go ahead with null values
#You have to deal with null values ->
#1. Remove the null values(delete)
#2. Fill the null values
```

```python
#1. Remove the null values(delete)
df_new = df.copy()
df_new
```

```python
df_new.dropna()          #dropna : drop the null values
df_new
```

```python
df_new.dropna(inplace=True)
df_new
```

```python
df_new.shape
```

```python
df_new.isnull().sum()
```

```python
#Fill the null value ->
#1. Fill with a value(default value)
#2. Fill with mean (mean value of the column)
#3.  Fill with median (median value of the column)
#4.  Fill with mode (mode value of the column)
```

```python
df.isnull().sum()
```

```python
df.fillna(0)           #Fill with any default value
# Temporary change (for permanent change -> inplace=True)
```

```python
df.fillna(df["qsec"].mean())   #Fill with mean value(average)
```

```python
df.fillna(df["qsec"].median())   #Fill with median value(middle)
```

```python
# Duplicate Values -> Duplicate Data (Repeated) 
# It is important to remove/drop the duplicate value and for that we use->
# df.drop_duplicates(inplace=True)
df.duplicated().sum()
```
