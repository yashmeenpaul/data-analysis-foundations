```python
#Line Plot ->
#Continuous data, Visualize the trend in data
#Basically to show the trend (how the data is flowing)
#plt.plot()
#plt.show() -> to show the data
```

```python
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.plot(days,sales)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Line Plot of Sales Per Day")
plt.show()
```

```python
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.plot(days,sales,color="Yellow",marker="o",linestyle="-.") 
 #marker is used to show the points (o,v,^,*), linestyle(dotted,dashed,etc.)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Line Plot of Sales Per Day")
plt.show()
```

```python
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.plot(days,sales,color="Yellow",marker="o",markersize=6,markerfacecolor = "red", markeredgecolor="blue",linestyle="-.") 
 #marker is used to show the points (o,v,^,*), linestyle(dotted,dashed,etc.)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Line Plot of Sales Per Day")
plt.grid()
plt.show()
```

```python
days = ["Mon","Tue","Wed","Thurs","Fri","Sat","Sun"]
mango_sales = [300,355,650,400,600,500,290]
grapes_sales = [400,305,400,420,800,200,590]
litchi_sales = [200,460,550,300,670,555,730]
plt.plot(days,mango_sales)
plt.plot(days,litchi_sales)
plt.plot(days,grapes_sales)
plt.show()
```

```python
days = ["Mon","Tue","Wed","Thurs","Fri","Sat","Sun"]
mango_sales = [300,355,650,400,600,500,290]
grapes_sales = [400,305,400,420,800,200,590]
litchi_sales = [200,460,550,300,670,555,730]
plt.plot(days,mango_sales,color="red",marker="h",markersize=6,markerfacecolor = "cyan",linestyle="dashed",label="Mango")
plt.plot(days,litchi_sales,color="black",marker="h",markersize=6,markerfacecolor = "cyan",linestyle="dashed",label="Lychee")
plt.plot(days,grapes_sales,color="pink",marker="h",markersize=6,markerfacecolor = "cyan", linestyle="dashed",label="Grapes")
plt.legend()
plt.show()
```

```python
#Bar Plot ->
# It is used to show the count of the Data and Compare b/w two Categories
```

```python
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.bar(days,sales,color=["pink","blue","yellow","red","orange","brown","green"])
plt.show()
```

```python
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.bar(days,sales,color=["pink","blue","yellow","red","orange","brown","green"],width=0.4)
plt.grid(color="black")
plt.show()
```

```python
#Pie Chart -> To show the proportion(percentage) of the categories related to whole data 
```

```python
value = [3,10,25,8,30]
categories = ["Mango","Apple","Lychee","Banana","Grapes"]
plt.pie(value,labels=categories, autopct="%0.0f%%",explode=[0.3,0,0,0,0])
plt.show()
```

```python
value = [3,10,25,8,30]
categories = ["Mango","Apple","Lychee","Banana","Grapes"]
plt.pie(value,labels=categories, autopct="%0.0f%%",explode=[0,0,0.8,0,0])
plt.show()
```

```python
value = [3,10,25,8,30]
categories = ["Mango","Apple","Lychee","Banana","Grapes"]
plt.pie(value,labels=categories, autopct="%0.0f%%",explode=[0,0.3,0.3,0,0],shadow='True')
plt.show()
```

```python
#Scatter Plot ->
#To visualize the spread of data
```

```python
roll_number = [1,2,3,4,5,6]
marks = [78,66,89,45,68,92]
plt.scatter(roll_number,marks,marker="*",s=200,color="black",edgecolor="pink")
plt.title("Marks of Students",color="red")
plt.xlabel("Roll Number",color="blue")
plt.ylabel("Marks",color="blue")
plt.show()
```

```python
roll_number = [1,2,3,4,5,6]
maths_marks = [78,66,89,45,68,92]
sci_marks = [88,56,79,95,88,42]
plt.scatter(roll_number,maths_marks,marker="*",s=200,color="black",edgecolor="pink",label="maths")
plt.scatter(roll_number,sci_marks,marker="^",s=200,color="black",edgecolor="blue",label="science")
plt.legend()
plt.title("Marks of Students",color="red")
plt.xlabel("Roll Number",color="blue")
plt.ylabel("Marks",color="blue")
plt.show()
```

```python
# Subplots ->
# When we are creating more than one plot together that is a subplot
```

```python
# 2 Plots Together ->
plt.subplots(1,2)

plt.subplot(1,2,1)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.plot(days,sales)

plt.subplot(1,2,2)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.bar(days,sales)

plt.show()
```

```python
# 4 plots together ->
plt.subplots(2,2)

plt.subplot(2,2,1)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.plot(days,sales)

plt.subplot(2,2,2)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.bar(days,sales)

plt.subplot(2,2,3)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.scatter(days,sales)

plt.subplot(2,2,4)
days = ["mon","tue","wed","thurs","fri","sat","sun"]
sales = [550,300,450,220,660,200,400]
plt.violinplot(sales)

plt.show()
```
